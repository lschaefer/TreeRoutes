```python
import pandas as pd
# local tools                                                                                                                                                                                             
import sys
import getImage
import matplotlib.pyplot as plt
```

```python
scenicDF = pd.read_csv('data/imagesWithJpg.tsv',sep='\t') # for debugging locally, update on aws
```

```python
scenicDF.info()
```

```python
ax = scenicDF.Average.plot.hist(bins=10,range=(1,11))
ax.set_xlabel('Scenery Score')
ax.set_ylabel('Counts')
plt.savefig('sceneryScore.png')
```

```python
ax = scenicDF.Variance.plot.hist()
ax.set_xlabel('Scenery Variance')
ax.set_ylabel('Counts')
plt.savefig('sceneryVariance.png')
```

```python
scenicDF["Nvotes"] = scenicDF['Votes'].apply(lambda x: len(x))

ax = scenicDF.Nvotes.plot.hist()
ax.set_yscale('log')
ax.set_xlabel('Number of votes')
ax.set_ylabel('Counts')
plt.savefig('nVotes.png')
```

```python
ax = scenicDF.Nvotes.plot.hist(range=(0,20))
ax.set_xlabel('Number of votes')
ax.set_ylabel('Counts')
plt.savefig('nVotes020.png')
```

```python
bestScore  = scenicDF.Average.idxmax()
worstScore = scenicDF.Average.idxmin()
```
